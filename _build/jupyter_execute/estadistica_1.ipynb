### ðŸ“Œ **Datos Cualitativos (CategÃ³ricos)**
Los datos cualitativos son aquellos que **describen atributos o caracterÃ­sticas** de un conjunto de elementos, sin involucrar valores numÃ©ricos que puedan ser operados matemÃ¡ticamente. Se utilizan para **clasificar** y **categorizar** la informaciÃ³n. Dentro de los datos cualitativos, encontramos dos subcategorÃ­as principales:

#### ðŸ”¹ **Datos Nominales**  
Son aquellos que simplemente **nombran o etiquetan** caracterÃ­sticas sin que exista un orden jerÃ¡rquico entre ellas. En estos datos, no se puede decir que una categorÃ­a es mayor o menor que otra.

âœ… **Ejemplos**:  
- GÃ©nero (Masculino, Femenino, No binario).  
- Tipo de automÃ³vil (SedÃ¡n, SUV, Hatchback).  
- PaÃ­s de origen (Colombia, Argentina, MÃ©xico).  
- Color de un producto (Rojo, Azul, Verde).  

> **ðŸ“Œ Nota**: Aunque algunas veces los datos nominales pueden representarse con nÃºmeros (Ejemplo: 1 para Rojo, 2 para Azul), esto no significa que puedan ser sometidos a operaciones matemÃ¡ticas.

#### ðŸ”¹ **Datos Ordinales**  
En este caso, las categorÃ­as **tienen un orden lÃ³gico o jerÃ¡rquico**, pero las diferencias entre los valores no son necesariamente uniformes o medibles.

âœ… **Ejemplos**:  
- Nivel de satisfacciÃ³n en una encuesta (Excelente, Bueno, Regular, Malo).  
- ClasificaciÃ³n de hoteles (1 estrella, 2 estrellas, 3 estrellasâ€¦).  
- TamaÃ±o de ropa (S, M, L, XL).  
- Nivel educativo alcanzado (Primaria, Secundaria, Universitario, Posgrado).  

> **ðŸ“Œ Importante**: Aunque los datos ordinales pueden ordenarse, **no se pueden realizar operaciones matemÃ¡ticas** directamente con ellos. Por ejemplo, la diferencia entre "Regular" y "Bueno" no es numÃ©ricamente equivalente a la diferencia entre "Bueno" y "Excelente".

---

### ðŸ”¢ **Datos Cuantitativos (NumÃ©ricos)**
Estos datos **representan cantidades o medidas** y permiten aplicar operaciones matemÃ¡ticas como sumas, restas, promedios, etc. Se dividen en:

#### ðŸ”¹ **Datos Cuantitativos Discretos**  
Son aquellos que **se obtienen mediante conteo** y solo pueden tomar **valores enteros**. No pueden fraccionarse en decimales o valores intermedios.

âœ… **Ejemplos**:  
- NÃºmero de empleados en una empresa.  
- Cantidad de productos vendidos en un dÃ­a.  
- NÃºmero de hijos en una familia.  
- Veces que un cliente compra en una tienda al mes.  

> **ðŸ“Œ CaracterÃ­stica clave**: Se obtienen por conteo y no admiten valores decimales. Por ejemplo, no podemos tener **2.5 empleados** en una empresa.

#### ðŸ”¹ **Datos Cuantitativos Continuos**  
Representan mediciones que pueden **tomar cualquier valor dentro de un rango determinado**, incluyendo valores fraccionarios o decimales.

âœ… **Ejemplos**:  
- Peso de un producto (500.3 gramos).  
- Altura de una persona (1.75 metros).  
- Precio de una acciÃ³n en la bolsa (34.78 USD).  
- Tiempo de espera en un servicio (2.35 minutos).  

> **ðŸ“Œ CaracterÃ­stica clave**: Se obtienen a travÃ©s de **mediciones** y pueden tomar valores decimales o fraccionarios.

---

### ðŸ“† **Â¿DÃ³nde ubicamos las fechas?**  
Las **fechas** son un tipo de dato particular que **puede clasificarse de diferentes maneras** dependiendo del contexto y el tipo de anÃ¡lisis que se realice.  

#### ðŸ”¹ **Fechas como datos cualitativos nominales**  
Cuando las fechas **se usan como etiquetas** sin un significado numÃ©rico, se consideran **cualitativas nominales**. 

âœ… **Ejemplos**:  
- AÃ±os de producciÃ³n de una empresa (2010, 2015, 2020, 2025).  
- Mes de nacimiento de los empleados (Enero, Febrero, Marzo).  
- DÃ­a de la semana con mÃ¡s ventas (Lunes, Martes, MiÃ©rcoles).  

> **ðŸ“Œ Nota**: En este caso, no se pueden realizar cÃ¡lculos como restas o promedios, ya que las fechas solo sirven como identificadores.

#### ðŸ”¹ **Fechas como datos cuantitativos continuos**  
Si se usan para **medir intervalos de tiempo**, como diferencias entre fechas o duraciÃ³n de eventos, se consideran **cuantitativas continuas**.

âœ… **Ejemplos**:  
- Tiempo en dÃ­as entre dos eventos importantes.  
- Edad de una persona (calculada a partir de la fecha de nacimiento).  
- Tiempo transcurrido desde el Ãºltimo mantenimiento de una mÃ¡quina.  
- DuraciÃ³n de un contrato en aÃ±os.  

> **ðŸ“Œ Importante**: Si las fechas se restan para encontrar intervalos, **se comportan como datos cuantitativos continuos**.

#### ðŸ”¹ **Fechas como datos ordinales**  
En algunos casos, si las fechas representan un orden, pero no se usan para cÃ¡lculos de intervalos, pueden tratarse como **datos ordinales**.

âœ… **Ejemplo**:  
- ClasificaciÃ³n de eventos por fecha (Primer, Segundo, Tercer lugar en una competencia).  
- Orden cronolÃ³gico de lanzamientos de productos.  

---

### ðŸ”„ **Â¿CÃ³mo elegir la mejor clasificaciÃ³n?**  
DependerÃ¡ de **cÃ³mo se usen los datos** en el anÃ¡lisis estadÃ­stico:

âœ… **Si las fechas sirven como etiquetas, son cualitativas nominales.**  
âœ… **Si se usan para medir diferencias de tiempo, son cuantitativas continuas.**  
âœ… **Si solo se ordenan cronolÃ³gicamente sin medir diferencias, pueden ser ordinales.**


## ðŸ“Œ **Ejemplo 1: Datos Cualitativos**
Vamos a trabajar con **datos nominales y ordinales**, creando un `DataFrame` y visualizÃ¡ndolos con grÃ¡ficos.

````python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creamos un DataFrame con datos cualitativos nominales y ordinales
df_cualitativo = pd.DataFrame({
    "Cliente": ["Carlos", "Ana", "Pedro", "SofÃ­a", "Luis"],
    "Color_Favorito": ["Rojo", "Azul", "Verde", "Azul", "Rojo"],  # Nominal
    "SatisfacciÃ³n": ["Bueno", "Excelente", "Regular", "Malo", "Bueno"]  # Ordinal
})

# Mapeo para convertir los datos ordinales a valores numÃ©ricos
orden_satisfaccion = {"Malo": 1, "Regular": 2, "Bueno": 3, "Excelente": 4}
df_cualitativo["SatisfacciÃ³n_Num"] = df_cualitativo["SatisfacciÃ³n"].map(orden_satisfaccion)

# Mostrar datos
display(df_cualitativo)

# VisualizaciÃ³n de datos cualitativos nominales con un conteo de colores favoritos
plt.figure(figsize=(6, 4))
sns.countplot(data=df_cualitativo, x="Color_Favorito", palette="Set2")
plt.title("DistribuciÃ³n de Colores Favoritos")
plt.xlabel("Color Favorito")
plt.ylabel("Frecuencia")
plt.show()

# VisualizaciÃ³n de datos ordinales con una barra de satisfacciÃ³n
plt.figure(figsize=(6, 4))
sns.barplot(data=df_cualitativo, x="Cliente", y="SatisfacciÃ³n_Num", palette="coolwarm")
plt.title("Nivel de SatisfacciÃ³n de Clientes")
plt.ylabel("Nivel de SatisfacciÃ³n (Ordinal)")
plt.xlabel("Cliente")
plt.show()
````

## ðŸ” **Inferencia EstadÃ­stica**
Hasta ahora hemos hablado de la **estadÃ­stica descriptiva**, que nos permite organizar y resumir datos. Sin embargo, muchas veces queremos **hacer conclusiones sobre una poblaciÃ³n sin analizar todos sus elementos**. AquÃ­ es donde entra la **inferencia estadÃ­stica**.

**ðŸ“Œ DefiniciÃ³n**  
La inferencia estadÃ­stica es el proceso de **extraer conclusiones sobre una poblaciÃ³n a partir de una muestra**. Se basa en la teorÃ­a de probabilidades y permite:

- **Estimaciones** de parÃ¡metros desconocidos (media, proporciÃ³n, varianza).  
- **Pruebas de hipÃ³tesis** para evaluar afirmaciones sobre una poblaciÃ³n.  
- **Modelos predictivos** para tomar decisiones con datos limitados.  

```{admonition} ðŸŽ¯ Ejemplo real
:class: tip
Supongamos que queremos saber la **edad promedio de los clientes de una tienda**. No podemos preguntar a todos, asÃ­ que tomamos una **muestra** y usamos inferencia estadÃ­stica para estimar la edad promedio de toda la poblaciÃ³n.
```

### ðŸ“Š **Ejemplo PrÃ¡ctico en Python: Inferencia EstadÃ­stica**
En este ejemplo, usaremos una **muestra aleatoria** para estimar la media poblacional.

````python
import numpy as np
import scipy.stats as stats

# Simulamos una poblaciÃ³n de edades de clientes (en aÃ±os)
np.random.seed(42)
poblacion_edades = np.random.normal(loc=35, scale=10, size=10000)  # Media 35, DesviaciÃ³n 10, 10,000 clientes

# Tomamos una muestra aleatoria de 100 clientes
muestra = np.random.choice(poblacion_edades, size=100, replace=False)

# Calculamos la media muestral y el intervalo de confianza del 95%
media_muestral = np.mean(muestra)
intervalo_confianza = stats.t.interval(0.95, len(muestra)-1, loc=media_muestral, scale=stats.sem(muestra))

# Resultados
print(f"ðŸ“Œ Media muestral estimada: {media_muestral:.2f} aÃ±os")
print(f"âœ… Intervalo de confianza del 95%: {intervalo_confianza}")
````

ðŸ“Œ **InterpretaciÃ³n**:  
El intervalo de confianza nos dice que, con un **95% de certeza**, la **media de la poblaciÃ³n** estÃ¡ dentro de ese rango. Si repitiÃ©ramos el experimento muchas veces, **el 95% de los intervalos calculados contendrÃ­an la media real**.
